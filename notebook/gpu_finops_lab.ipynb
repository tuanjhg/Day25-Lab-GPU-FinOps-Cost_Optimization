```python
# Cell 17: Time-series cost tracking
print("📈 Collecting time-series data (10 snapshots)...")
ts_data = []
for i in range(10):
    snap = requests.post(f"{GATEWAY_URL}/cost/snapshot").json()
    ts_data.append({
        'time': i,
        'total_cost': snap['total_cost_usd'],
        'idle_cost': snap['total_idle_cost_usd'],
        'active_cost': snap['total_active_cost_usd'],
        'waste_pct': snap['waste_pct'],
    })
    time.sleep(1)

df = pd.DataFrame(ts_data)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Cost over time
ax1.stackplot(df['time'], df['active_cost'], df['idle_cost'],
              labels=['Active (productive)', 'Idle (waste)'],
              colors=['#4CAF50', '#F44336'], alpha=0.8)
ax1.set_xlabel('Time (intervals)')
ax1.set_ylabel('Cost per interval (USD)')
ax1.set_title('GPU Cost Allocation Over Time')
ax1.legend(loc='upper left')

# Waste percentage
ax2.plot(df['time'], df['waste_pct'], 'r-o', linewidth=2)
ax2.axhline(y=30, color='orange', linestyle='--', label='Warning threshold (30%)')
ax2.axhline(y=50, color='red', linestyle='--', label='Critical threshold (50%)')
ax2.set_xlabel('Time (intervals)')
ax2.set_ylabel('Waste %')
ax2.set_title('GPU Cost Waste Percentage')
ax2.legend()
ax2.set_ylim(0, 100)

plt.tight_layout()
plt.savefig('finops_timeseries.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Part 7: Complete FinOps Workflow
Run a full cycle: submit workloads → monitor → detect waste → autoscale → optimize cost.

```python
# Cell 18: Full FinOps Optimization Workflow
print("🔄 FULL FINOPS OPTIMIZATION WORKFLOW")
print("=" * 60)

# Step 1: Check initial state
print("\n1️⃣  Initial cluster state:")
m = requests.get(f"{GATEWAY_URL}/cluster/metrics").json()
print(f"   GPUs: {m['total_gpus']} | Util: {m['avg_utilization']:.1f}% | Idle: {m['idle_gpus']}")

# Step 2: Submit heavy workloads to increase utilization
print("\n2️⃣  Submitting heavy workloads...")
for i in range(5):
    requests.post(f"{GATEWAY_URL}/cluster/workloads/submit", json={
        "workload_id": f"heavy-{i:03d}", "gpu_type_preferred": "T4", "gpu_count": 1, "duration_seconds": 600
    })
m = requests.get(f"{GATEWAY_URL}/cluster/metrics").json()
print(f"   After load: Util: {m['avg_utilization']:.1f}% | Busy: {m['busy_gpus']}/{m['total_gpus']}")

# Step 3: Autoscaler evaluates
print("\n3️⃣  Autoscaler evaluation:")
decision = requests.post(f"{GATEWAY_URL}/autoscaler/evaluate").json()
print(f"   Decision: {decision['action']} - {decision['reason']}")

# Step 4: Cost snapshot
print("\n4️⃣  Cost analysis:")
snap = requests.post(f"{GATEWAY_URL}/cost/snapshot").json()
print(f"   Total cost/interval: ${snap['total_cost_usd']:.6f}")
print(f"   Waste: {snap['waste_pct']:.1f}%")

# Step 5: Get recommendations
print("\n5️⃣  Recommendations:")
recs = requests.post(f"{GATEWAY_URL}/cost/recommendations").json()
for rec in recs:
    print(f"   [{rec['priority']}] {rec['type']}: savings ~{rec['estimated_savings_pct']}%")

# Step 6: Apply optimization - use spot for some workloads
print("\n6️⃣  Applying optimization: Switch to spot instances...")
for i in range(3):
    requests.post(f"{GATEWAY_URL}/spot/request", json={
        "instance_id": f"opt-spot-{i}", "gpu_type": "T4",
        "gpu_count": 1, "max_price_per_hour": 0.15, "workload_id": f"heavy-{i:03d}"
    })

savings = requests.get(f"{GATEWAY_URL}/spot/savings-report").json()
print(f"   Spot savings: ${savings['total_savings']:.4f} ({savings['savings_pct']:.1f}%)")

# Step 7: Complete workloads and record billing
print("\n7️⃣  Final billing:")
for i in range(5):
    requests.post(f"{GATEWAY_URL}/cluster/workloads/heavy-{i:03d}/complete")
    requests.post(f"{GATEWAY_URL}/billing/record", json={
        "workload_id": f"heavy-{i:03d}", "gpu_type": "T4",
        "gpu_count": 1, "duration_seconds": 600, "is_spot": i < 3
    })

final = requests.get(f"{GATEWAY_URL}/billing/summary").json()
print(f"   Total spend: ${final['total_cost_usd']:.4f}")
print(f"   Total saved: ${final['total_savings_usd']:.4f}")
print(f"   Budget: {final['budget_utilization_pct']:.1f}% used")
print(f"\n✅ Workflow complete!")
```

```python
# Cell 20: GPU Metrics Collection (works on Kaggle/Colab)
# Strategy: pynvml first (reliable Python lib), torch.cuda fallback

def get_gpu_metrics():
    """Get GPU metrics inline. Called during training loop."""

    # Method 1: pynvml (most reliable on Kaggle - pure Python, no subprocess)
    if PYNVML_AVAILABLE:
        try:
            nvmlInit()
            handle = nvmlDeviceGetHandleByIndex(0)
            util = nvmlDeviceGetUtilizationRates(handle)
            mem_info = nvmlDeviceGetMemoryInfo(handle)
            try:
                power = nvmlDeviceGetPowerUsage(handle) / 1000.0  # mW -> W
            except Exception:
                power = 0.0
            try:
                temp = nvmlDeviceGetTemperature(handle, 0)  # NVML_TEMPERATURE_GPU = 0
            except Exception:
                temp = 0.0
            return {
                'timestamp': time.time(),
                'gpu_util_pct': float(util.gpu),
                'mem_util_pct': float(util.memory),
                'mem_used_mb': mem_info.used / 1e6,
                'mem_total_mb': mem_info.total / 1e6,
                'power_watts': float(power),
                'temp_c': float(temp),
            }
        except Exception as e:
            pass  # Fall through to torch.cuda

    # Method 2: torch.cuda (always works, limited metrics)
    mem_used = torch.cuda.memory_allocated() / 1e6
    mem_reserved = torch.cuda.memory_reserved() / 1e6
    mem_total = torch.cuda.get_device_properties(0).total_memory / 1e6
    util_approx = (mem_reserved / mem_total) * 100 if mem_total > 0 else 0
    return {
        'timestamp': time.time(),
        'gpu_util_pct': util_approx,
        'mem_util_pct': util_approx,
        'mem_used_mb': mem_used,
        'mem_total_mb': mem_total,
        'power_watts': 0.0,
        'temp_c': 0.0,
    }

# === DIAGNOSTIC TEST ===
print("=" * 50)
print("GPU METRICS DIAGNOSTIC")
print("=" * 50)

# Test pynvml
print(f"\n1. pynvml available: {PYNVML_AVAILABLE}")
if PYNVML_AVAILABLE:
    try:
        nvmlInit()
        handle = nvmlDeviceGetHandleByIndex(0)
        util = nvmlDeviceGetUtilizationRates(handle)
        mem = nvmlDeviceGetMemoryInfo(handle)
        print(f"   pynvml works! GPU util={util.gpu}%, mem={mem.used/1e6:.0f}/{mem.total/1e6:.0f} MB")
        try:
            pwr = nvmlDeviceGetPowerUsage(handle) / 1000.0
            print(f"   Power: {pwr:.1f}W")
        except Exception as e:
            print(f"   Power: not available ({e})")
        try:
            tmp = nvmlDeviceGetTemperature(handle, 0)
            print(f"   Temp: {tmp}C")
        except Exception as e:
            print(f"   Temp: not available ({e})")
    except Exception as e:
        print(f"   pynvml FAILED: {e}")

# Test full function
print(f"\n2. get_gpu_metrics() test:")
test_result = get_gpu_metrics()
for k, v in sorted(test_result.items()):
    if k != 'timestamp':
        print(f"   {k}: {v}")

print(f"\n   Method: {'pynvml' if test_result.get('temp_c', 0) > 0 or (PYNVML_AVAILABLE and test_result.get('gpu_util_pct', -1) >= 0) else 'torch.cuda'}")
print(f"\nReady for training.")
```

```python
# Cell 21: Prepare CIFAR-10 dataset and ResNet-18 model
transform_train = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=128, shuffle=True, num_workers=2)

testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
]))
testloader = torch.utils.data.DataLoader(testset, batch_size=256, shuffle=False, num_workers=2)

print(f"Dataset ready: {len(trainset)} train / {len(testset)} test images")
print(f"Batches per epoch: {len(trainloader)}")


def create_model():
    """Create a fresh ResNet-18 for CIFAR-10."""
    model = torchvision.models.resnet18(weights=None, num_classes=10)
    model.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1, bias=False)
    model.maxpool = nn.Identity()
    return model.cuda()


def train_epoch_monitored(model, loader, optimizer, criterion, use_amp=False, sample_interval=10):
    """Train one epoch with inline GPU metrics collection every N batches."""
    model.train()
    total_loss, correct, total = 0, 0, 0
    scaler = GradScaler() if use_amp else None
    metrics_samples = []
    start = time.time()

    for batch_idx, (inputs, targets) in enumerate(loader):
        inputs, targets = inputs.cuda(), targets.cuda()
        optimizer.zero_grad()

        if use_amp:
            with autocast():
                outputs = model(inputs)
                loss = criterion(outputs, targets)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

        total_loss += loss.item() * inputs.size(0)
        _, predicted = outputs.max(1)
        correct += predicted.eq(targets).sum().item()
        total += inputs.size(0)

        # Collect GPU metrics every N batches (inline, no threading)
        if batch_idx % sample_interval == 0:
            torch.cuda.synchronize()
            sample = get_gpu_metrics()
            metrics_samples.append(sample)

    elapsed = time.time() - start
    return total_loss / total, correct / total * 100, elapsed, metrics_samples


print("Model & training functions ready (inline monitoring every 10 batches)")
```

```python
# Cell 22: Train FP32 (baseline) with inline GPU monitoring
NUM_EPOCHS = 3
price_per_hour = GPU_PRICING[detected_type]

print("=" * 60)
print("EXPERIMENT 1: FP32 Training (Baseline)")
print("=" * 60)

model_fp32 = create_model()
optimizer_fp32 = optim.SGD(model_fp32.parameters(), lr=0.1, momentum=0.9, weight_decay=5e-4)
criterion = nn.CrossEntropyLoss()

torch.cuda.synchronize()
torch.cuda.reset_peak_memory_stats()
fp32_start = time.time()

fp32_results = []
fp32_all_metrics = []
for epoch in range(NUM_EPOCHS):
    loss, acc, t, epoch_metrics = train_epoch_monitored(model_fp32, trainloader, optimizer_fp32, criterion, use_amp=False)
    fp32_results.append({"epoch": epoch+1, "loss": loss, "accuracy": acc, "time_s": t})
    fp32_all_metrics.extend(epoch_metrics)
    print(f"   Epoch {epoch+1}/{NUM_EPOCHS} | Loss: {loss:.4f} | Acc: {acc:.1f}% | Time: {t:.1f}s | Samples: {len(epoch_metrics)}")

fp32_total_time = time.time() - fp32_start
fp32_peak_mem = torch.cuda.max_memory_allocated() / 1e9
fp32_cost = (fp32_total_time / 3600) * price_per_hour

print(f"\n   Total samples collected: {len(fp32_all_metrics)}")

# Build monitor dataframe
fp32_monitor_df = pd.DataFrame(fp32_all_metrics)
if not fp32_monitor_df.empty and 'timestamp' in fp32_monitor_df.columns:
    fp32_monitor_df['elapsed'] = fp32_monitor_df['timestamp'] - fp32_monitor_df['timestamp'].iloc[0]

print(f"\n   FP32 Summary:")
print(f"   Total time:      {fp32_total_time:.1f}s")
print(f"   Peak memory:     {fp32_peak_mem:.2f} GB")
if not fp32_monitor_df.empty and 'gpu_util_pct' in fp32_monitor_df.columns:
    print(f"   Avg GPU util:    {fp32_monitor_df['gpu_util_pct'].mean():.1f}%")
    print(f"   Avg power:       {fp32_monitor_df['power_watts'].mean():.1f}W")
    print(f"   Avg temperature: {fp32_monitor_df['temp_c'].mean():.1f}C")
    print(f"   Max GPU util:    {fp32_monitor_df['gpu_util_pct'].max():.1f}%")
else:
    print(f"   GPU util:        DataFrame empty! columns={list(fp32_monitor_df.columns) if not fp32_monitor_df.empty else 'EMPTY'}")
    if fp32_all_metrics:
        print(f"   First sample keys: {list(fp32_all_metrics[0].keys())}")
print(f"   Estimated cost:  ${fp32_cost:.6f}")
```

```python
# Cell 27: Multi-GPU Cost Analysis
print("=" * 60)
print("EXERCISE 8.5.1: Multi-GPU Cost Analysis")
print("=" * 60)

def analyze_multi_gpu_cost(base_time_hours, gpu_type, gpu_counts, scaling_factors=None):
    """
    Analyze training time and cost across multiple GPU counts.
    """
    pricing = globals().get(
        "GPU_PRICING",
        {"T4": 0.35, "A100": 3.67, "V100": 2.48, "P100": 1.46, "L4": 0.81},
    )
    if gpu_type not in pricing:
        raise ValueError(f"Unknown gpu_type: {gpu_type}")
    price_per_hour = pricing[gpu_type]

    if scaling_factors is None:
        scaling_factors = {}
        for n in gpu_counts:
            speedup = n ** 0.85
            scaling_factors[n] = min(speedup, n * 0.95)

    results = []
    for n in sorted(gpu_counts):
        speedup = max(scaling_factors.get(n, n ** 0.85), 1e-6)
        time_hours = base_time_hours / speedup
        total_cost = time_hours * price_per_hour * n
        efficiency = speedup / n
        cost_per_speedup = total_cost / speedup
        results.append(
            {
                "gpu_count": n,
                "speedup": speedup,
                "efficiency": efficiency,
                "time_hours": time_hours,
                "total_cost": total_cost,
                "cost_per_speedup": cost_per_speedup,
            }
        )

    best = min(results, key=lambda r: r["cost_per_speedup"])
    return {
        "gpu_type": gpu_type,
        "price_per_hour": price_per_hour,
        "base_time_hours": base_time_hours,
        "results": results,
        "best": best,
    }


# Test data
base_training_time = 2.0  # hours for single GPU
test_gpu_counts = [1, 2, 4, 8]

analysis = analyze_multi_gpu_cost(
    base_training_time,
    detected_type if "detected_type" in globals() else "T4",
    test_gpu_counts,
)

df = pd.DataFrame(analysis["results"])
df["efficiency_pct"] = df["efficiency"] * 100
df["time_hours"] = df["time_hours"].round(3)
df["total_cost"] = df["total_cost"].round(4)
df["cost_per_speedup"] = df["cost_per_speedup"].round(4)
df = df[
    [
        "gpu_count",
        "speedup",
        "efficiency_pct",
        "time_hours",
        "total_cost",
        "cost_per_speedup",
    ]
]

print("\nMulti-GPU analysis results:")
display(df)

best = analysis["best"]
print(
    f"\nBest cost efficiency: {best['gpu_count']} GPUs "
    f"(cost_per_speedup=${best['cost_per_speedup']:.4f}, "
    f"time={best['time_hours']:.2f}h, total_cost=${best['total_cost']:.2f})"
)
```

```python
# Cell 28: Project Cost Forecasting
print("=" * 60)
print("EXERCISE 8.5.2: Project Cost Forecasting")
print("=" * 60)

import math


def forecast_project_cost(phases, contingency_pct=20, confidence_level=0.95):
    """
    Forecast multi-phase project costs with uncertainty and confidence intervals.
    """
    pricing = globals().get(
        "GPU_PRICING",
        {"T4": 0.35, "A100": 3.67, "V100": 2.48, "P100": 1.46, "L4": 0.81},
    )

    breakdown = []
    total_base = 0.0
    variance = 0.0

    for phase in phases:
        gpu_type = phase["gpu_type"]
        if gpu_type not in pricing:
            raise ValueError(f"Unknown gpu_type: {gpu_type}")

        price = pricing[gpu_type]
        duration = phase["duration_hours"]
        gpu_count = phase["gpu_count"]
        uncertainty = float(phase.get("uncertainty_pct", 0.0))

        base_cost = price * gpu_count * duration
        std = base_cost * uncertainty

        total_base += base_cost
        variance += std**2

        breakdown.append(
            {
                "name": phase["name"],
                "gpu_type": gpu_type,
                "gpu_count": gpu_count,
                "duration_hours": duration,
                "base_cost": base_cost,
                "uncertainty_pct": uncertainty,
                "std_cost": std,
            }
        )

    contingency_amount = total_base * (contingency_pct / 100.0)
    total_expected = total_base + contingency_amount

    z_map = {0.90: 1.64, 0.95: 1.96, 0.99: 2.58}
    z = z_map.get(round(confidence_level, 2), 1.96)
    total_std = math.sqrt(variance)

    ci_low = max(0.0, total_base - z * total_std) + contingency_amount
    ci_high = (total_base + z * total_std) + contingency_amount

    return {
        "phases": breakdown,
        "total_base_cost": total_base,
        "contingency_amount": contingency_amount,
        "total_expected_cost": total_expected,
        "confidence_level": confidence_level,
        "ci_low": ci_low,
        "ci_high": ci_high,
    }


# Example project phases
example_project = [
    {
        "name": "Data Preparation",
        "gpu_type": "T4",
        "gpu_count": 1,
        "duration_hours": 40,
        "uncertainty_pct": 0.15,
    },
    {
        "name": "Model Training",
        "gpu_type": "A100",
        "gpu_count": 4,
        "duration_hours": 120,
        "uncertainty_pct": 0.25,
    },
    {
        "name": "Hyperparameter Tuning",
        "gpu_type": "A100",
        "gpu_count": 8,
        "duration_hours": 60,
        "uncertainty_pct": 0.30,
    },
    {
        "name": "Model Evaluation",
        "gpu_type": "T4",
        "gpu_count": 2,
        "duration_hours": 20,
        "uncertainty_pct": 0.10,
    },
]

forecast = forecast_project_cost(example_project, contingency_pct=20, confidence_level=0.95)

phase_df = pd.DataFrame(forecast["phases"])
phase_df["base_cost"] = phase_df["base_cost"].round(2)
phase_df["std_cost"] = phase_df["std_cost"].round(2)
phase_df["uncertainty_pct"] = (phase_df["uncertainty_pct"] * 100).round(1)
phase_df = phase_df[
    [
        "name",
        "gpu_type",
        "gpu_count",
        "duration_hours",
        "base_cost",
        "uncertainty_pct",
        "std_cost",
    ]
]

print("\nPhase cost breakdown:")
display(phase_df)

print("\nForecast summary:")
print(f"Total base cost:       ${forecast['total_base_cost']:.2f}")
print(f"Contingency buffer:    ${forecast['contingency_amount']:.2f}")
print(f"Expected total cost:   ${forecast['total_expected_cost']:.2f}")
print(
    f"{int(forecast['confidence_level']*100)}% CI: "
    f"${forecast['ci_low']:.2f} to ${forecast['ci_high']:.2f}"
)
```

```python
# Cell 29: Optimization Opportunity Analysis
print("=" * 60)
print("EXERCISE 8.5.3: Advanced Optimization Opportunity Analysis")
print("=" * 60)


def analyze_optimization_opportunities(current_config, optimization_strategies):
    """
    Analyze and prioritize optimization strategies by savings, effort, and risk.
    """
    pricing = globals().get(
        "GPU_PRICING",
        {"T4": 0.35, "A100": 3.67, "V100": 2.48, "P100": 1.46, "L4": 0.81},
    )
    if current_config["gpu_type"] not in pricing:
        raise ValueError(f"Unknown gpu_type: {current_config['gpu_type']}")

    base_cost = (
        pricing[current_config["gpu_type"]]
        * current_config["gpu_count"]
        * current_config["duration_hours"]
    )

    effort_score_map = {"LOW": 1, "MEDIUM": 2, "HIGH": 3}
    risk_penalty_map = {"LOW": 0.0, "MEDIUM": 0.15, "HIGH": 0.35}

    available_names = {s["name"] for s in optimization_strategies}

    ranked = []
    for s in optimization_strategies:
        effort = effort_score_map.get(s.get("implementation_effort", "MEDIUM"), 2)
        risk_level = s.get("risk_level", "MEDIUM")
        risk_penalty = risk_penalty_map.get(risk_level, 0.15)
        savings_pct = float(s.get("savings_pct", 0.0))
        adjusted_savings = savings_pct * (1 - risk_penalty)
        score = adjusted_savings / effort if effort > 0 else 0.0

        deps = s.get("dependencies", [])
        blocked = any(dep not in available_names for dep in deps)

        ranked.append(
            {
                "name": s["name"],
                "savings_pct": savings_pct,
                "adjusted_savings_pct": adjusted_savings,
                "implementation_effort": s.get("implementation_effort", "MEDIUM"),
                "effort_score": effort,
                "risk_level": risk_level,
                "dependencies": deps,
                "blocked": blocked,
                "score": score,
                "expected_savings_usd": base_cost * adjusted_savings,
            }
        )

    ranked.sort(key=lambda r: (r["blocked"], -r["score"]))

    cumulative_factor = 1.0
    for r in ranked:
        if r["blocked"]:
            r["cumulative_savings_pct"] = None
            continue
        cumulative_factor *= (1 - r["adjusted_savings_pct"])
        r["cumulative_savings_pct"] = 1 - cumulative_factor

    quick_wins = [
        r
        for r in ranked
        if not r["blocked"] and r["effort_score"] <= 1 and r["risk_level"] == "LOW"
    ]

    return {
        "base_cost": base_cost,
        "ranked": ranked,
        "quick_wins": quick_wins,
    }


# Example optimization strategies
example_strategies = [
    {
        "name": "Switch to Mixed Precision (AMP)",
        "savings_pct": 0.25,
        "implementation_effort": "LOW",
        "risk_level": "LOW",
        "dependencies": [],
    },
    {
        "name": "Use Spot Instances",
        "savings_pct": 0.60,
        "implementation_effort": "MEDIUM",
        "risk_level": "HIGH",
        "dependencies": [],
    },
    {
        "name": "Optimize Batch Size",
        "savings_pct": 0.15,
        "implementation_effort": "LOW",
        "risk_level": "LOW",
        "dependencies": [],
    },
    {
        "name": "Implement Early Stopping",
        "savings_pct": 0.20,
        "implementation_effort": "MEDIUM",
        "risk_level": "LOW",
        "dependencies": [],
    },
    {
        "name": "Switch to More Efficient GPU Type",
        "savings_pct": 0.40,
        "implementation_effort": "HIGH",
        "risk_level": "MEDIUM",
        "dependencies": [],
    },
]

current_training_config = {
    "gpu_type": "A100",
    "gpu_count": 4,
    "duration_hours": 100,
    "precision": "FP32",
    "instance_type": "on-demand",
}

opt = analyze_optimization_opportunities(current_training_config, example_strategies)

rank_df = pd.DataFrame(opt["ranked"])
rank_df["savings_pct"] = (rank_df["savings_pct"] * 100).round(1)
rank_df["adjusted_savings_pct"] = (rank_df["adjusted_savings_pct"] * 100).round(1)
rank_df["expected_savings_usd"] = rank_df["expected_savings_usd"].round(2)
rank_df["score"] = rank_df["score"].round(3)
rank_df = rank_df[
    [
        "name",
        "savings_pct",
        "adjusted_savings_pct",
        "implementation_effort",
        "risk_level",
        "expected_savings_usd",
        "score",
        "blocked",
    ]
]

print("\nPrioritized optimization strategies:")
display(rank_df)

if opt["quick_wins"]:
    print("\nQuick wins (low effort, low risk):")
    for r in opt["quick_wins"]:
        print(f"- {r['name']}")
```

```python
# Cell 30: Integrated Cost Dashboard
print("=" * 60)
print("EXERCISE 8.5.4: Integrated Cost Dashboard")
print("=" * 60)


def create_advanced_finops_dashboard(
    multi_gpu_analysis, project_forecast, optimization_recommendations
):
    """
    Create a multi-panel dashboard with Part 8.5 analyses.
    """
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))

    # Plot 1: Multi-GPU cost curve
    results = multi_gpu_analysis["results"]
    counts = [r["gpu_count"] for r in results]
    costs = [r["total_cost"] for r in results]
    axes[0, 0].plot(counts, costs, marker="o", color="#1f77b4")
    axes[0, 0].set_title("Total Cost vs GPU Count")
    axes[0, 0].set_xlabel("GPU Count")
    axes[0, 0].set_ylabel("Total Cost (USD)")

    # Plot 2: Scaling efficiency
    efficiency = [r["efficiency"] for r in results]
    axes[0, 1].plot(counts, efficiency, marker="o", color="#2ca02c")
    axes[0, 1].set_title("Scaling Efficiency")
    axes[0, 1].set_xlabel("GPU Count")
    axes[0, 1].set_ylabel("Speedup / GPU")
    axes[0, 1].set_ylim(0, 1.05)

    # Plot 3: Project forecast with confidence interval
    expected = project_forecast["total_expected_cost"]
    ci_low = project_forecast["ci_low"]
    ci_high = project_forecast["ci_high"]
    axes[0, 2].bar(["Expected"], [expected], color="#ff7f0e")
    axes[0, 2].errorbar(
        ["Expected"],
        [expected],
        yerr=[[expected - ci_low], [ci_high - expected]],
        fmt="none",
        ecolor="black",
        capsize=6,
    )
    axes[0, 2].set_title("Project Cost Forecast")
    axes[0, 2].set_ylabel("Cost (USD)")

    # Plot 4: Phase cost breakdown
    phase_names = [p["name"] for p in project_forecast["phases"]]
    phase_costs = [p["base_cost"] for p in project_forecast["phases"]]
    axes[1, 0].bar(phase_names, phase_costs, color="#9467bd")
    axes[1, 0].set_title("Phase Cost Breakdown")
    axes[1, 0].set_ylabel("Cost (USD)")
    axes[1, 0].tick_params(axis="x", rotation=30)

    # Plot 5: Optimization prioritization (savings vs effort)
    ranked = optimization_recommendations["ranked"]
    effort_map = {"LOW": 1, "MEDIUM": 2, "HIGH": 3}
    x_effort = [effort_map.get(r["implementation_effort"], 2) for r in ranked]
    y_savings = [r["savings_pct"] * 100 for r in ranked]
    axes[1, 1].scatter(x_effort, y_savings, color="#d62728")
    axes[1, 1].set_title("Savings vs Effort")
    axes[1, 1].set_xlabel("Effort (1=LOW, 3=HIGH)")
    axes[1, 1].set_ylabel("Savings (%)")

    # Plot 6: Cumulative savings roadmap
    roadmap = [r for r in ranked if not r["blocked"]]
    cum = []
    cum_factor = 1.0
    for r in roadmap:
        cum_factor *= (1 - r["adjusted_savings_pct"])
        cum.append((1 - cum_factor) * 100)
    axes[1, 2].plot(range(1, len(cum) + 1), cum, marker="o", color="#17becf")
    axes[1, 2].set_title("Cumulative Savings Roadmap")
    axes[1, 2].set_xlabel("Number of Strategies")
    axes[1, 2].set_ylabel("Cumulative Savings (%)")

    plt.tight_layout()
    return fig


# Use results from earlier exercises
if "analysis" not in globals():
    analysis = analyze_multi_gpu_cost(2.0, "T4", [1, 2, 4, 8])
if "forecast" not in globals():
    forecast = forecast_project_cost(example_project)
if "opt" not in globals():
    opt = analyze_optimization_opportunities(current_training_config, example_strategies)

fig = create_advanced_finops_dashboard(analysis, forecast, opt)
plt.savefig("advanced_finops_dashboard.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: advanced_finops_dashboard.png")
```

```python
# Cell 31: Challenge Exercise - Cost Optimization Strategy Design
print("=" * 60)
print("CHALLENGE EXERCISE 8.5.5: Design Cost Optimization Strategy")
print("=" * 60)

# Scenario input
challenge_scenario = {
    "project": "Large Language Model Fine-tuning",
    "requirements": {
        "training_duration": 200,  # hours baseline
        "gpu_type": "A100",
        "gpu_count": 8,
        "precision": "FP32",
        "instance_type": "on-demand",
        "budget": 5000,  # USD
    },
    "constraints": {
        "max_preemption_risk": "MEDIUM",
        "min_accuracy": 0.95,
        "deadline": "2 weeks",
    },
}

print("Scenario:")
print(f"Project: {challenge_scenario['project']}")
print(
    f"Baseline: {challenge_scenario['requirements']['gpu_count']}x "
    f"{challenge_scenario['requirements']['gpu_type']} for "
    f"{challenge_scenario['requirements']['training_duration']}h"
)
print(f"Budget: ${challenge_scenario['requirements']['budget']}")
print(f"Deadline: {challenge_scenario['constraints']['deadline']}")

pricing = globals().get(
    "GPU_PRICING",
    {"T4": 0.35, "A100": 3.67, "V100": 2.48, "P100": 1.46, "L4": 0.81},
)

req = challenge_scenario["requirements"]
constraints = challenge_scenario["constraints"]

gpu_type = req["gpu_type"]
if gpu_type not in pricing:
    raise ValueError(f"Unknown gpu_type: {gpu_type}")

baseline_cost = pricing[gpu_type] * req["gpu_count"] * req["training_duration"]
print(f"\nBaseline cost: ${baseline_cost:.2f}")

# Estimate single-GPU equivalent time from baseline configuration
baseline_speedup = req["gpu_count"] ** 0.85
base_time_hours = req["training_duration"] * baseline_speedup

candidate_counts = sorted(set([1, 2, 4, 8, req["gpu_count"]]))
analysis = analyze_multi_gpu_cost(base_time_hours, gpu_type, candidate_counts)

# Parse deadline to hours
import re

deadline_text = constraints.get("deadline", "")
deadline_hours = None
week_match = re.search(r"(\d+)\s*week", deadline_text, re.I)
day_match = re.search(r"(\d+)\s*day", deadline_text, re.I)
if week_match:
    deadline_hours = int(week_match.group(1)) * 7 * 24
elif day_match:
    deadline_hours = int(day_match.group(1)) * 24

candidates = analysis["results"]
if deadline_hours is not None:
    candidates = [c for c in candidates if c["time_hours"] <= deadline_hours]

chosen = min(candidates, key=lambda r: r["total_cost"]) if candidates else analysis["best"]
print(
    f"\nChosen GPU count: {chosen['gpu_count']} GPUs "
    f"(time={chosen['time_hours']:.1f}h, total_cost=${chosen['total_cost']:.2f})"
)

# Filter strategies by max allowed risk
risk_rank = {"LOW": 1, "MEDIUM": 2, "HIGH": 3}
max_risk = constraints.get("max_preemption_risk", "HIGH")
allowed_risk = risk_rank.get(max_risk, 3)

filtered_strategies = [
    s
    for s in example_strategies
    if risk_rank.get(s.get("risk_level", "MEDIUM"), 2) <= allowed_risk
]

current_config = {
    "gpu_type": gpu_type,
    "gpu_count": chosen["gpu_count"],
    "duration_hours": chosen["time_hours"],
    "precision": req.get("precision", "FP32"),
    "instance_type": req.get("instance_type", "on-demand"),
}

opt = analyze_optimization_opportunities(current_config, filtered_strategies)
ranked = opt["ranked"]
selected = ranked[:3] if ranked else []

cumulative_factor = 1.0
for s in selected:
    cumulative_factor *= (1 - s["adjusted_savings_pct"])
cumulative_savings_pct = 1 - cumulative_factor

# Forecast final cost with uncertainty and apply cumulative savings
phase = [
    {
        "name": "Training",
        "gpu_type": gpu_type,
        "gpu_count": chosen["gpu_count"],
        "duration_hours": chosen["time_hours"],
        "uncertainty_pct": 0.25,
    }
]
forecast = forecast_project_cost(phase, contingency_pct=15, confidence_level=0.95)

final_expected = forecast["total_expected_cost"] * (1 - cumulative_savings_pct)
final_ci_low = forecast["ci_low"] * (1 - cumulative_savings_pct)
final_ci_high = forecast["ci_high"] * (1 - cumulative_savings_pct)

print("\nSelected strategies:")
for s in selected:
    print(
        f"- {s['name']} (adj savings {s['adjusted_savings_pct']*100:.1f}%, "
        f"effort {s['implementation_effort']}, risk {s['risk_level']})"
    )

print("\nFinal cost forecast after optimizations:")
print(f"Expected: ${final_expected:.2f}")
print(f"95% CI:  ${final_ci_low:.2f} to ${final_ci_high:.2f}")

budget = req.get("budget")
if budget is not None:
    status = "OK" if final_expected <= budget else "OVER"
    print(f"Budget check: {status} (budget ${budget:.2f})")
```
